# Load a custom dataset 

Вы можете попробовать загрузить свой собственный датасет с Yandex Drive.

```python
import yadisk
import zipfile

!mkdir -p data/datasets
dataset_link = "https://disk.360.yandex.ru/d/Mzh9T9OEZKSsww" # insert your link

y = yadisk.YaDisk()
y.download_public(dataset_link, "dataset.zip")

with zipfile.ZipFile("dataset.zip", 'r') as zip_ref:
    zip_ref.extractall("data/user_data")
```
